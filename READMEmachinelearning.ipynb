# Credit Score Prediction Using Regression Models

## Project Overview
This project involves building and comparing multiple regression models to predict credit scores based on estimated salary and account balance. The goal is to explore different machine learning techniques and identify the best model for accurate credit score prediction.

## Dataset
The dataset includes financial data such as estimated salary and balance, along with corresponding credit scores.

## Models Used
- Multiple Linear Regression  
- Random Forest Regression  
- Support Vector Regression (SVR)  
- Polynomial Regression  
- Lasso Regression  

## Tools & Libraries
- Python  
- pandas  
- scikit-learn  
- matplotlib  

## Project Steps
1. Data preprocessing and cleaning  
2. Exploratory data analysis  
3. Feature engineering  
4. Model building and training  
5. Hyperparameter tuning  
6. Model evaluation and comparison  
7. Results visualization
